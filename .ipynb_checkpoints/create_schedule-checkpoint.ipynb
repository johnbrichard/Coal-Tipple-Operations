```python
import numpy as np
import pandas as pd
```

```python
schedule = pd.DataFrame(np.random.randint(5,21,size=(10000, 3)), columns=['TrainA','TrainB','TrainC'])
```

```python
schedule.insert(loc=len(schedule.columns),column='TrainHC',value=np.random.randint(11,14,size=(10000,1)))
```

```python
schedule.head()
```

```python
for x in range(0,len(schedule['TrainHC'])):
    if x%7==4:
        pass
    else:
        schedule['TrainHC'][x]=-1
```

```python
schedule.head()
```

### Check for duplicate arrival times for trains. 
### If two times coincide, generate random time for one.

```python
for x in range(0,len(schedule['TrainHC'])):
    a=schedule.loc[x,'TrainA']
    b=schedule.loc[x,'TrainB']
    c=schedule.loc[x,'TrainC']
    hc=schedule.loc[x,'TrainHC']
    
    if x%7==4:
        record_set=[a,b,c,hc]
        for y in range(0,len(record_set)):
            for z in range(y+1,len(record_set)):
                if record_set[y]==record_set[z]:
                    if z==(len(record_set)-1):
                        record_set[z]=np.random.randint(11,14)
                    else:
                        record_set[z]=np.random.randint(5,21)
            
        schedule.loc[x,'TrainA']=record_set[0]
        schedule.loc[x,'TrainB']=record_set[1]
        schedule.loc[x,'TrainC']=record_set[2]
        schedule.loc[x,'TrainHC']=record_set[3]
        
    else:
        record_set=[a,b,c]
        for y in range(0,len(record_set)):
            for z in range(y+1,len(record_set)):
                if record_set[y]==record_set[z]:
                    record_set[z]=np.random.randint(5,21)
        
        schedule.loc[x,'TrainA']=record_set[0]
        schedule.loc[x,'TrainB']=record_set[1]
        schedule.loc[x,'TrainC']=record_set[2]
```
